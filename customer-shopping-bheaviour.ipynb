```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_ammount'})
```

```python
df.columns
```

```python
labels = ['young-adult', 'adult','middle-age','senior']
df['age_groups'] = pd.qcut(df['age'],q=4,labels=labels)
```

```python
frequency_mapping = {
'Fortnightly': 14,
'Weekly': 7,
'Monthly': 30,
'Quarterly': 90,
'Bi-Weekly': 14,
'Annually': 365,
'Every 3 Months': 90

}
df['purchase_frequency_dates'] = df['frequency_of_purchases'].map(frequency_mapping)
```
